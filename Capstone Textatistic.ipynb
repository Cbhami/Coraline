```python
from textatistic import Textatistic
from nltk import sent_tokenize
import syllables

# Open and read the plain text file
with open('/Users/cole/Library/CloudStorage/OneDrive-Personal/Coraline/DCL_308_HamiltonC_Capstone.txt', 'r', encoding='latin-1') as file:
    text_content = file.read()

# Create an instance of Textatistic
textatistic = Textatistic(text_content)

# Split the text into sentences and count them
sentences = sent_tokenize(text_content)
sentence_count = len(sentences)

# Access other attributes if desired
char_count = textatistic.char_count
word_count = textatistic.word_count

# Calculate average sentence length and average syllables per word
avg_sentence_length = len(text_content) / sentence_count
avg_syllables_per_word = sum([syllables.estimate(w) for w in text_content.split()]) / word_count

# Calculate Flesch Reading Ease score
flesch_reading_ease_score = 206.835 - (1.015 * avg_sentence_length) - (84.6 * avg_syllables_per_word)

# Print the obtained insights
print("Text Statistics")
print("---------------")
print(f"Character count:      {char_count:,}")
print(f"Word count:          {word_count:,}")
print(f"Sentence count:      {sentence_count:,}\n")

# Calculate and print readability metrics
print("Readability Metrics")
print("-------------------")
flesch_score = textatistic.flesch_score
fleschkincaid_score = textatistic.fleschkincaid_score
gunningfog_score = textatistic.gunningfog_score
smog_score = textatistic.smog_score

print(f"Flesch score:        {flesch_score:.2f}")
print(f"Flesch-Kincaid score: {fleschkincaid_score:.2f}")
#print(f"Syllable count:      {syllable_count:,}")
print(f"Gunningfog score:    {gunningfog_score:.2f}")
print(f"SMOG score:          {smog_score:.2f}")
print(f"Flesch Reading Ease: {flesch_reading_ease_score:.2f}")

# Print the results
print("\nResults")
print("-------")
print(f"Flesch Reading Ease:     {flesch_reading_ease_score:.2f}")
print(f"Flesch-Kincaid Grade Level: {fleschkincaid_score:.2f}")
print(f"Gunning Fog Index:        {gunningfog_score:.2f}")
print(f"SMOG Index:               {smog_score:.2f}")

# Print the definitions of these scores
print("\nDefinitions")
print("-----------")
print("Flesch Reading Ease: The higher the score, the easier the text is to read.")
print("Flesch-Kincaid Grade Level: The grade level at which the text is readable.")
print("Gunning Fog Index: Indicates the years of education required to understand the text.")
print("SMOG Index: Indicates the years of education required to understand the text.")

# Narrative Summary
print("\nNarrative Summary")
print("-----------------")
print("The paper has the following statistics:")
print(f"- The total character count is {char_count:,}.")
print(f"- The total word count is {word_count:,}.")
print(f"- The total sentence count is {sentence_count:,}.")

print("The readability metrics indicate:")
print(f"- The Flesch Reading Ease score is {flesch_reading_ease_score:.2f}.")
print(f"- The Flesch-Kincaid Grade Level is {fleschkincaid_score:.2f}.")
print(f"- The Gunning Fog Index is {gunningfog_score:.2f}.")
print(f"- The SMOG Index is {smog_score:.2f}.")
```
