```r
plot1 <- ggplot(df, aes(x = price, y = month)) + geom_point() + geom_smooth(method = "lm") + labs(title = "Price vs Date", x = "Date", y = "Price")
```

```r
plot1
```

```r
plot2 <- ggplot(df, aes(x = date)) + geom_bar() + labs(title = "Number of Orders by Date", x = "Date", y = "Number of Orders")
```

```r
plot2
```

```r
plot3 <- ggplot(df, aes(x = date)) + geom_density() + labs(title = "Density of Orders by Date", x = "Date", y = "Density")
```

```r
plot3
```

```r
df$day <- wday(df$date, label = TRUE)
df$week <- week(df$date)
df$weekday <- wday(df$date, label = TRUE, abbr = TRUE)
df$month <- month(df$date, label = TRUE)
df$year <- year(df$date)
df$hour <- hour(df$time)
```

```r
plot4 <- ggplot(df, aes(x = day, y= price
   )) + geom_boxplot() + labs(title = "Boxplot of Price by Date", x = "Date", y = "Price", group = "day")
```

```r
plot4
```

```r
plot5 <- ggplot(df, aes(sample = price)) + geom_qq() + labs(title = "QQ Plot of Price", x = "Theoretical Quantiles", y = "Sample Quantiles") + geom_qq_line()
```

```r
plot5
```
