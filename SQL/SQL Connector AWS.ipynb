```python
#read the table "Activity_Workouts" from awsdb2 INTO A DATAFRAME
cursor.execute("SELECT * FROM Activity_Workouts")
result = cursor.fetchall()
df = pd.DataFrame(result, columns=['id', 'Start_Time', 'End_Time', 'Session_Type', 'Session_Subtype', 'Creation_Method', 'Local_Time_Zone_Offset', 'Activity_Score', 'Intense_Duration_ms', 'Moderate_Duration_ms', 'Max_Heart_Rate_bpm', 'Average_Heart_Rate_bpm', 'Steps', 'Calories', 'Pace_ms', 'Distance_m', 'Elevation_m', 'Software_Version', 'Software_OS', 'Software_OS_Version', 'Active_Time_ms', 'Total_Duration_ms', 'Activity_Pause', 'Session_Source'])
df.head()
```

```python
df.info()
```

```python
#convert Start_Time and End_Time to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['End_Time'] = pd.to_datetime(df['End_Time'])
df.head()
```

```python
#Convert "No Data" to 0 in the entire dataframe
df = df.replace('No Data', 0)
#convert the columns 'Intense_Duration_ms', 'Moderate_Duration_ms', 'Max_Heart_Rate_bpm', 'Average_Heart_Rate_bpm', 'Steps', 'Calories', 'Pace_ms', 'Distance_m', 'Elevation_m', 'Active_Time_ms', 'Total_Duration_ms' to numeric
df['Intense_Duration_ms'] = pd.to_numeric(df['Intense_Duration_ms'])
df['Moderate_Duration_ms'] = pd.to_numeric(df['Moderate_Duration_ms'])
df['Max_Heart_Rate_bpm'] = pd.to_numeric(df['Max_Heart_Rate_bpm'])
df['Average_Heart_Rate_bpm'] = pd.to_numeric(df['Average_Heart_Rate_bpm'])
df['Steps'] = pd.to_numeric(df['Steps'])
df['Calories'] = pd.to_numeric(df['Calories'])
df['Pace_ms'] = pd.to_numeric(df['Pace_ms'])
df['Distance_m'] = pd.to_numeric(df['Distance_m'])
df['Elevation_m'] = pd.to_numeric(df['Elevation_m'])
df['Active_Time_ms'] = pd.to_numeric(df['Active_Time_ms'])
df['Total_Duration_ms'] = pd.to_numeric(df['Total_Duration_ms'])
df.info()
```

```python
#separate date and time from Start_Time and End_Time
df['Start_Date'] = df['Start_Time'].dt.date
df['Start_Time'] = df['Start_Time'].dt.time
df['End_Date'] = df['End_Time'].dt.date
df['End_Time'] = df['End_Time'].dt.time
df.head()
```
